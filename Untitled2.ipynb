```python
import cv2
import numpy as np
import cv2 as cv
import pandas as pd
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from PIL import Image
from skimage import io

def detect_colors(image):
    # Convert image to HSV color space
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define color ranges
    colors = {
        'red': ([0, 50, 50], [10, 255, 255]),
        'blue': ([100, 50, 50], [130, 255, 255]),
        'green': ([40, 50, 50], [80, 255, 255]),
        # Add more color ranges as needed
    }

    # Detect colors in the image
    color_results = {}
    for color, (lower, upper) in colors.items():
        lower = np.array(lower, dtype=np.uint8)
        upper = np.array(upper, dtype=np.uint8)
        mask = cv2.inRange(hsv_image, lower, upper)
        color_results[color] = cv2.countNonZero(mask)

    return color_results

# Load image from file
# Replace with the path to your image file
image = io.imread("/content/8678740.jpg")

# Detect colors in the image
color_results = detect_colors(image)

# Print the color results
for color, count in color_results.items():
    print(f"{color}: {count}")

# Display the image
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
